## Data Description

- Survival : 0 = No, 1 = Yes
- pclass(Ticket Class) : 1 = 1st, 2 = 2nd, 3 = 3rd
- Sex(Gender) : Male, Female
- Age : Age in years
- SibSp : Number of siblings/spouses abroad the titanic
- Parch : Number of parents/children abrod the titanic
- Ticket : Ticket Number
- Fare : Passenger fare
- Cabin : Cabin Number
- Embarked : Port of Embarkation, C = Cherbourg, Q = Queenstown, S = Southampton

# Import Seaborn

```python
import seaborn as sns
```

# Q9: Draw barplot represents survived people based on sex

```python
sns.barplot(x=df['sex'], y=df['survived'], estimator=np.mean)
```

# Q10: Draw catplot represents survived people based on embarked

```python
sns.catplot(
    x = 'survived',
    data=df,
    kind='count',
    col='embarked'
)
```

# Q11: Draw boxplot represents distribution of male and female based on age and pclass

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,8))
sns.boxplot(
    x="sex", 
    y="age",
    hue ='pclass',
    data=df)
```

# Q12: Draw heatmap represents correlations between sibsp, parch, age, fare, and survived columns

```python
corr_matrix = df[['sibsp', 'parch', 'age', 'fare', 'survived']].corr()

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='RdYlGn',
    vmin=-1,
    vmax=1
)

plt.title('Correlation Heatmap for Titanic Passenger Features')
plt.xlabel('Feature')
plt.ylabel('Feature')

plt.show()
```

# Q13: Draw factorplot represents the relation between sibsp and survived columns

```python
sns.catplot(
    x = 'sibsp',
    y = 'survived',kind='point',
    data=df
)

plt.show()
```
